```python
import geopandas as gpd
import plotly.express as px

# Carregue o shapefile
# shapefile_inundavel = "SAD69-96_SHP_areainundavel_polygon.shp"
shapefile_inundavel = "SIRGAS_SHP_areainundavel_polygon.shp"
gdf_inundavel = gpd.read_file(shapefile_inundavel)

# Verifique e defina o CRS inicial (SIRGAS UTM Zona 23S)
if gdf_inundavel.crs is None:
    gdf_inundavel = gdf_inundavel.set_crs(epsg=31983)

# Reproje para WGS84 (EPSG:4326)
gdf_inundavel = gdf_inundavel.to_crs(epsg=4326)

# Crie o mapa com Plotly
fig = px.choropleth_mapbox(
    gdf_inundavel,
    geojson=gdf_inundavel.geometry,
    locations=gdf_inundavel.index,
    color_discrete_sequence=["lightblue"],  # Cor personalizada
    mapbox_style="carto-positron",
    center=dict(lat=-23.55, lon=-46.63),  # Centro aproximado
    zoom=9
)
fig.update_traces(marker_line_width=0.5, marker_line_color="black")  # Adiciona bordas
fig.show()
```

```python
gdf_inundavel.to_file('inundavel.geojson', driver='GeoJSON', encoding='utf-8')
```
