```python
import geopandas as gpd
import os

shapefile_favela = 'DEINFO_FAVELAS_2015.shp'
distritos = 'distritos.geojson'
gdf_favela = gpd.read_file(shapefile_favela)
gdf_distritos = gpd.read_file(distritos)

gdf_favela
```

```python
gdf_distritos
```

```python
# encode favela's geometry to SP
gdf_favela = gdf_favela.to_crs(epsg=4326)

gdf_favela
```

```python
# check if favelas are inside distritos
gdf_favela.intersects(gdf_distritos.geometry)

# add distrito label to favelas if more than 50% of the area is inside the distrito
gdf_favela['DISTRITO'] = gdf_favela.geometry.apply(lambda x: gdf_distritos[gdf_distritos.geometry.intersects(x)].iloc[0]['NOME_DIST'] if gdf_distritos.geometry.intersects(x).any() else None)

gdf_favela
```

```python

# calculate its area
gdf_favela[gdf_favela['DISTRITO'].isna()].area

# drop it
gdf_favela = gdf_favela.dropna(subset=['DISTRITO'])

gdf_favela
```

```python
# simplify geometry
gdf_favela['geometry'] = gdf_favela['geometry'].simplify(tolerance=0.0005)

```

```python
import plotly.express as px

fig = px.choropleth_mapbox(gdf_favela, geojson=gdf_favela.geometry, locations=gdf_favela.index, mapbox_style="carto-positron", center=dict(lat=-23.55, lon=-46.63), zoom=9)
fig.show()
```

```python
# save to geojson with all columns
gdf_favela.to_file('favelas_final.geojson', driver='GeoJSON', encoding='utf-8')

# verify columns in saved file
import geopandas as gpd
saved_gdf = gpd.read_file('favelas_final.geojson')
print("Columns in saved file:", saved_gdf.columns)
```
