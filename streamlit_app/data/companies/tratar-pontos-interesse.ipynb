```python
import pandas as pd

paths = {
    "Academia de Ginastica": "Academia de Ginastica.xlsx",
    "Agencia Bancaria": "Agencia Bancaria.xlsx",
    "Escola Privada": "Escola Privada.xlsx",
    "Escola Pública": "Escola Pública.xlsx",
    "Faculdade" : "faculdade.xlsx",
    "Farmacia": "Farmacia.xlsx",
    "Fast Food": "Fast Food.xlsx",
    "Hipermercados e Supermercados": "Hipermercados e Supermercados.xlsx",
    "Hortifruti": "Hortifruti.xlsx",
    "Hospital": "Hospital.xlsx",
    "Petshop": "Petshop.xlsx",
    "Restaurante": "Restaurantes.xlsx",
    "Shopping Center": "Shopping Center.xlsx",

}

dfs = {k: pd.read_excel(v) for k, v in paths.items()}

# filter municipio to be São Paulo everywhere
dfs = {k: v[v["Município"] == "São Paulo"] for k, v in dfs.items()}
```

```python

# drop Unnamed columns, Telecone, Municipio, Estado, Mesorregião	Região Geográfica	Microárea, Telefone
for k, v in dfs.items():
    v.drop(v.columns[v.columns.str.contains('unnamed', case=False)], axis=1, inplace=True, errors="ignore")
    v.drop(["Município", "Estado", "Mesorregião", "Microrregião", "Região Geográfica", "Microárea", "Telefone"], axis=1, inplace=True, errors="ignore")
    # make first column the id column and change it to id
    v.rename(columns={v.columns[0]: "id"}, inplace=True)
    v.set_index("id", inplace=True)
```

```python
# iterate over all dataframes
for k, v in dfs.items():
    # if there is a column with "Nome da Rede", change it to "Rede"
    if "Nome da Rede" in v.columns:
        v.rename(columns={"Nome da Rede": "Rede"}, inplace=True)
    # if there is a column with the name in the list, change it to "Nome"
    list = ["Nome", "Nome da Unidade", "Nome Instituição"]
    for l in list:
        if l in v.columns:
            v.rename(columns={l: "Nome"}, inplace=True)
```

```python
# display dfs heads
for k, v in dfs.items():
    display(k)
    display(v.head())
    display(v.shape)
```

```python
# add the first and last 5 letters of the name to the id
for k, v in dfs.items():
    v.index = v.index.map(lambda x: f"{k}_{x}")
```

```python
# create a Tipo column with the key
for k, v in dfs.items():
    v["Tipo"] = k
```

```python
dfs["Academia de Ginastica"].head()
```

```python
# keep only Rede	Nome	Localizado em	Endereço	Bairro	CEP	Tipo
for k, v in dfs.items():
    v.drop([c for c in v.columns if c not in ["Rede", "Nome", "Localizado em", "Endereço", "Bairro", "CEP", "Tipo"]], axis=1, inplace=True)
    # if there is no Rede column, create it null
    if "Rede" not in v.columns:
        v["Rede"] = None
```

```python
# print all cols to check
for k, v in dfs.items():
    # sort ['Rede', 'Nome', 'Localizado em', 'Endereço', 'Bairro', 'CEP', 'Tipo']
    v = v[['Rede', 'Nome', 'Localizado em', 'Endereço', 'Bairro', 'CEP', 'Tipo']]
    dfs[k] = v
```

```python
# print heads
for k, v in dfs.items():
    display(k)
    display(v.head())
    display(v.shape)
```

```python
import pandas as pd
import requests

# Carrega o arquivo existente com dados parciais
name = "Escola Privada"
df = pd.read_csv(f"{name}.csv")

# Token e endpoint do Mapbox
token = "pk.eyJ1IjoicHJvamV0b2RhZG9zIiwiYSI6ImNtMXdiNjVobTBpa2Eya3BsMnR5OWxsd3AifQ.SGh5qTES1kmMN3VNFzZAwQ"
SAO_PAULO = (-46.63389, -23.55028)

# Loop para obter coordenadas de cada endereço apenas para as linhas que estão sem Latitude e Longitude
count = 0
for i, row in df.iterrows():
    # Verifica se a latitude e longitude já estão preenchidas
    if pd.notna(row['Latitude']) and pd.notna(row['Longitude']):
        continue  # Pula esta linha se as coordenadas já estão presentes

    # Monta o texto de busca e a URL para a API
    search_text = f"{row['Endereço']} {row['Bairro']} São Paulo"
    url = f"https://api.mapbox.com/geocoding/v5/mapbox.places/{search_text}.json?access_token={token}&proximity={SAO_PAULO[0]},{SAO_PAULO[1]}"

    try:
        # Faz a solicitação à API
        response = requests.get(url)
        data = response.json()

        # Verifica se há dados nas features
        if not data.get("features"):
            print(f"Nenhum resultado para o endereço: {search_text}")
            continue

        # Atualiza o DataFrame com a nova Latitude e Longitude
        df.at[i, "Latitude"] = data["features"][0]["geometry"]["coordinates"][1]
        df.at[i, "Longitude"] = data["features"][0]["geometry"]["coordinates"][0]

        # Incrementa o contador e exibe o progresso a cada 100 registros
        count += 1
        if count % 100 == 0:
            print(f"{count} registros processados")

        # Salva o progresso a cada 1000 linhas processadas
        if count % 1000 == 0:
            df.to_csv(f"{name}.csv", index=False)

    except requests.exceptions.RequestException as e:
        print(f"Erro na solicitação para o endereço {search_text}: {e}")

# Salva o arquivo final após processar todas as linhas
df.to_csv(f"{name}.csv", index=False)
print("Processo concluído e dados salvos em 'Escola Privada.csv'")
```
